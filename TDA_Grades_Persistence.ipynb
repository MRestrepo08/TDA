```python
from persim import plot_diagrams
```

```python
import teaspoon.MakeData.PointCloud as makePtCloud
import teaspoon.TDA.Draw as Draw
```

```python
df1 = pd.read_excel('19_1_1.xlsx', sheet_name='Hoja_1')
df2 = pd.read_excel('20_1_1.xlsx', sheet_name='Hoja_1')
df3 = pd.read_excel('21_1_1.xlsx', sheet_name='Hoja_1')
```

```python
df1['NOMBRE_ASIGNATURA'] = df1['NOMBRE_ASIGNATURA'].map(asignaturas2)
df2['NOMBRE_ASIGNATURA'] = df2['NOMBRE_ASIGNATURA'].map(asignaturas2)
df3['NOMBRE_ASIGNATURA'] = df3['NOMBRE_ASIGNATURA'].map(asignaturas2)
```

```python
data1 = df1[(df1['CONTRATO'] == 3) & (df1['ESTAAPROCALI'] == 0)]
data2 = df2[(df2['CONTRATO'] == 3) & (df2['ESTAAPROCALI'] == 0)]
data3 = df3[(df3['CONTRATO'] == 3) & (df3['ESTAAPROCALI'] == 0)]
```

```python
features = [c for c in df1.columns if c not in ['Unnamed: 0','NOMBRE_PROFESOR','NOMBRE_ASIGNATURA']]
```

```python
X1 = np.array(data1[features].fillna(0))
X2 = np.array(data2[features].fillna(0))
X3 = np.array(data3[features].fillna(0))
```

```python
diagrams1 = ripser.ripser(X1)['dgms']
plot_diagrams(diagrams1, show=True)
```

```python
diagrams2 = ripser.ripser(X2)['dgms']
plot_diagrams(diagrams2, show=True)
```

```python
diagrams3 = ripser.ripser(X3)['dgms']
plot_diagrams(diagrams3, show=True)
```

```python
Draw.drawDgm(diagrams1[0])
Draw.drawDgm(diagrams2[0])
Draw.drawDgm(diagrams3[0])
```

```python
Draw.drawDgm(diagrams1[1])
Draw.drawDgm(diagrams2[1])
Draw.drawDgm(diagrams3[1])
```

```python
list = [diagrams1[0],diagrams2[0],diagrams3[0]]
distancias = []
for _ in list:
    dist, matching = persim.bottleneck(diagrams1[0], _, matching=True)
    distancias.append(dist)
print(distancias)
```

```python
list = [diagrams1[1],diagrams2[1],diagrams3[1]]
distancias = []
for _ in list:
    dist, matching = persim.bottleneck(diagrams1[1], _, matching=True)
    distancias.append(dist)
print(distancias)
```
